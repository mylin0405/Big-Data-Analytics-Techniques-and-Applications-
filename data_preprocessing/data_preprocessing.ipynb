```python
# read data: "BDA_final.csv"
df = pd.read_csv("df_all_stocks.csv")
df
```

```python
# drop the column "Unnamed: 0"
df = df.drop(columns = ["Unnamed: 0"])
```

```python
# I want to see the frequency of "tic"
fre_cnt = df["tic"].value_counts()
tmp = fre_cnt.to_frame()
pd.set_option('display.max_columns', None)
tmp.head(100)
```

```python
df = df.sort_values(by=["tic", "date"])
```

```python
df_tickets = df.groupby("tic")
for idx, df_ticket in df_tickets:
    #print(idx)
    #print(df_ticket)
    df_ticket.to_csv(f"./data/df_{idx}.csv", index=False)
    
```
